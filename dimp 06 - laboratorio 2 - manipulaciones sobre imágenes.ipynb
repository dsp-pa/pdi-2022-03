### Transformaciones bsicas sobre imgenes

#### Inversin

```python
A = Image.open('images/lab02.webp')
A = np.asarray(A)
```

```python
L = 256
I = L - 1 - A

plt.imshow(I, cmap='gray')
```

#### Suma y resta

```python
A = Image.open('images/lab02.webp').convert('L')
A = np.asarray(A)

B = np.full_like(A, 64)

C = A +  B

Image.fromarray(C).show()
plt.imshow(C, cmap='gray', vmin=0, vmax=255)
```

```python
A = Image.open('images/lab02.webp').convert('L')
A = np.asarray(A)

B = np.full_like(A, 64)

C = A - B

Image.fromarray(C).show()
plt.imshow(C, cmap='gray', vmin=0, vmax=255)
```

#### Transformaciones logartmicas y exponenciales

```python
A = Image.open('images/lab02.webp').convert('L')
A = np.asarray(A)

plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(A, cmap='gray', vmin=A.min(), vmax=A.max())

B = np.log(1 + A) / np.log(2)

plt.subplot(2, 2, 2)
plt.imshow(B, cmap='gray', vmin=B.min(), vmax=B.max())

B = np.log(1 + A) / np.log(5)

plt.subplot(2, 2, 3)
plt.imshow(B, cmap='gray', vmin=B.min(), vmax=B.max())

B = np.log(1 + A) / np.log(10)

plt.subplot(2, 2, 4)
plt.imshow(B, cmap='gray', vmin=B.min(), vmax=B.max())
```
